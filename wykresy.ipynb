```python
file_name = 'message_1.json'
path_to_file = 'G:\Moje\Programowanie\Projekty_Data\Python\Messenger Analysis\main analysis\Data\magdalenaszanter_1583136002126624'
full_path = f'{path_to_file}\{file_name}'
full_path
```

```python
import json

# Wczytaj dane z pliku JSON
with open(full_path, 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
# Zainicjuj liczniki wiadomości od każdej osoby
message_count = {}
total_messages = len(data['messages'])

# Przejdź przez każdą wiadomość i zlicz ilość od każdej osoby
for message in data['messages']:
    sender_name = message['sender_name']
    message_count[sender_name] = message_count.get(sender_name, 0) + 1

# Oblicz procentowy udział wiadomości od każdej osoby
percentage_messages = {}
for sender, count in message_count.items():
    percentage = (count / total_messages) * 100
    percentage_messages[sender] = percentage

# Wydrukuj procentowy udział wiadomości od każdej osoby
for sender, percentage in percentage_messages.items():
    print(f"{sender}: {percentage:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Przykładowe dane dotyczące procentowego udziału w 4 okresach
okresy = [1, 2, 3, 4]  # Numery okresów
procenty_magdaleny = [59.73, 62.86, 58.54, 55.43]  # Procentowy udział Magdaleny w każdym okresie
procenty_szymona = [40.27, 37.14, 41.46, 44.57]  # Procentowy udział Szymona w każdym okresie

# Tworzenie wykresu z dostosowanym wyglądem
plt.figure(figsize=(8, 6))
plt.style.use('seaborn-whitegrid')  # Zmiana stylu tła na jasne tło siatkowe

# Linie dla Magdaleny z okrągłymi markerami i niebieskim kolorem
plt.plot(okresy, procenty_magdaleny, marker='o', color='blue', label='Magdalena', linewidth=2)

# Linie dla Szymona z trójkątnymi markerami i czerwonym kolorem
plt.plot(okresy, procenty_szymona, marker='^', color='red', label='Szymon', linewidth=2)

# Dodanie etykiet i tytułu
plt.xlabel('Okres')
plt.ylabel('Procentowy udział')
plt.title('Procentowy udział Magdaleny i Szymona w 4 okresach')
plt.xticks(okresy)

# Legenda
plt.legend()

# Wyświetlenie wykresu
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Przykładowe dane dotyczące procentowego udziału w 4 okresach
okresy = [1, 2, 3, 4]  # Numery okresów
procenty_magdaleny = [59.73, 62.86, 58.54, 55.43]  # Procentowy udział Magdaleny w każdym okresie
procenty_szymona = [40.27, 37.14, 41.46, 44.57]  # Procentowy udział Szymona w każdym okresie

# Tworzenie wykresu słupkowego z dostosowanym wyglądem
plt.figure(figsize=(8, 6))
width = 0.35  # Szerokość słupków

# Słupki dla Magdaleny w kolorze niebieskim
plt.bar([okres - width/2 for okres in okresy], procenty_magdaleny, width, label='Magdalena', color='blue', alpha=0.7)

# Słupki dla Szymona w kolorze czerwonym
plt.bar([okres + width/2 for okres in okresy], procenty_szymona, width, label='Szymon', color='red', alpha=0.7)

# Dodanie etykiet i tytułu
plt.xlabel('Okres')
plt.ylabel('Procentowy udział')
plt.title('Procentowy udział Magdaleny i Szymona w 4 okresach')
plt.xticks(okresy)

# Legenda
plt.legend()

# Usunięcie ramki
plt.box(False)

# Dodanie wartości nad słupkami
for x, y1, y2 in zip(okresy, procenty_magdaleny, procenty_szymona):
    plt.text(x - width/2, y1 + 1, f'{y1:.2f}%', ha='center', va='bottom', color='blue', fontsize=10)
    plt.text(x + width/2, y2 + 1, f'{y2:.2f}%', ha='center', va='bottom', color='red', fontsize=10)

# Wyświetlenie wykresu
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
